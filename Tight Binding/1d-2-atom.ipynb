# Tight Binding
## Trans-polyacetylene
### Let $\phi_s$ be a s wave function
### One can construct a Bloch wave, $\psi$, on a periodic lattice

$\Phi_{i \mathbf{k}}(\mathbf{r})=\frac{1}{\sqrt{N}} \sum_{\mathbf{R}_{i}} e^{i \mathbf{k} \cdot \mathbf{R}_{i}} \phi\left(\mathbf{r}-\mathbf{R}_{i}\right)$

### Matrix elements are calculated as following

$H_{i j}=\left\langle\Phi_{i \mathbf{k}}|H| \Phi_{j \mathbf{k}}\right\rangle$

### To find the energy eigenvalues, we must diagonalize the system

$|\mathbf{H}-E(\mathbf{k}) \mathbf{I}|=0$



## 2-Atom Basis

$\left|\begin{array}{cc}H_{\mathrm{AA}}-E & H_{\mathrm{AB}} \\ H_{\mathrm{BA}} & H_{\mathrm{BB}}-E\end{array}\right|=0$

### where $H_{\mathrm{AB}}=H_{\mathrm{BA}}^{*}$

### As a result, dispersion relation becomes,

$E(\mathbf{k})=-\frac{1}{2}\left(H_{\mathrm{AA}}+H_{\mathrm{BB}}\right) \pm \sqrt{\frac{1}{4}\left(H_{\mathrm{AA}}-H_{\mathrm{BB}}\right)^{2}+\left|H_{\mathrm{AB}}\right|^{2}}$

Where $H_{AA}$, $H_{AB}$, $H_{BA}$ ande $H_{BB}$ are the following [[Reading Material]](trans-polyacetylene)



## $\mathbf{R}_{\mathrm{A}}=\pm a \mathbf{i}$, $\mathbf{R}_{\mathrm{B}}=\mathbf{R}_{\mathrm{AB}} \pm a \mathbf{i}$, $\mathbf{R}_{\mathrm{AB}}=(a / 2, a / 2 \sqrt{3})$


$\begin{aligned} H_{\mathrm{AA}} &=\frac{1}{N} \sum_{\mathbf{R}_{\mathrm{A}}} \sum_{\mathbf{R}_{\mathrm{A}}^{\prime}} e^{i \mathbf{k} \cdot\left(\mathbf{R}_{\mathrm{A}}^{\prime}-\mathbf{R}_{\mathrm{A}}\right)} \int \phi_{s}^{*}\left(\mathbf{r}-\mathbf{R}_{\mathrm{A}}\right) H \phi_{s}\left(\mathbf{r}-\mathbf{R}_{\mathrm{A}}^{\prime}\right) d \mathbf{r} \\ &=\sum_{\mathbf{R}_{\mathrm{A}}^{\prime \prime}} e^{i \mathbf{k} \cdot \mathbf{R}_{\mathrm{A}}^{\prime \prime}} \int \phi_{s}^{*}(\mathbf{x}) H \phi_{s}\left(\mathbf{x}-\mathbf{R}_{\mathrm{A}}^{\prime \prime}\right) d \mathbf{x} \\ &=\epsilon_{p}+\sum_{m \neq 0} e^{i m k a} \gamma(|m a|) \end{aligned}$

$H_{\mathrm{AA}}=H_{\mathrm{BB}}=\epsilon_{p}+2 \gamma(a) \cos (k a)$

$\begin{aligned} H_{\mathrm{AB}} &=\frac{1}{N} \sum_{\mathbf{R}_{\mathrm{A}}} \sum_{\mathbf{R}_{\mathrm{B}}} e^{i \mathbf{k} \cdot\left(\mathbf{R}_{\mathrm{A}}-\mathbf{R}_{\mathrm{B}}\right)} \int \phi_{s}^{*}\left(\mathbf{r}-\mathbf{R}_{\mathrm{A}}\right) H \phi_{s}\left(\mathbf{r}-\mathbf{R}_{\mathrm{B}}\right) d \mathbf{r}, \\ &=\sum_{\mathbf{R}_{\mathrm{A}}{ }^{\prime}} e^{i \mathbf{k} \cdot\left(\mathbf{R}_{\mathrm{AB}}+\mathbf{R}_{\mathrm{A}}^{\prime}\right)} \int \phi_{s}^{*}(\mathbf{x}) H \phi_{s}\left(\mathbf{x}-\left(\mathbf{R}_{\mathrm{AB}}+\mathbf{R}_{\mathrm{A}}^{\prime}\right)\right) d \mathbf{x} \end{aligned}$

$\begin{aligned} H_{\mathrm{AB}} &=\left(e^{i k a / 2}+e^{-i k a / 2}\right) \gamma(|\boldsymbol{\tau}|) \\ &=2 \cos (k a / 2) \gamma(|\boldsymbol{\tau}|) \end{aligned}$


## Finally, E becomes,

### $E(k)=\epsilon_{p}+2 \gamma(a) \cos (k a) \pm 2 \cos (k a / 2) \gamma(|\boldsymbol{\tau}|)$
